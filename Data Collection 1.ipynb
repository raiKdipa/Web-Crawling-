```python
import csv
import requests
from bs4 import BeautifulSoup

# URL of the webpage
url = 'https://privacy.microsoft.com/en-gb/privacystatement'

# Send a GET request to the webpage
response = requests.get(url)

# Create BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find the target parent element using CSS selectors
parent_element = soup.select_one('div#main[name="main"][role="main"]')

# Find all the desired elements (h1, h2, p, li) within the parent element
elements = parent_element.select('h1, h2, p, li')

# Print the text attribute and save the parsed data to a CSV file
with open('microsoft_policy_data_2023.csv', 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Text'])
    for element in elements:
        text = element.get_text(strip=True)
        print(text)
        writer.writerow([text])
```

```python
import csv
import requests
from bs4 import BeautifulSoup

# Website URL
url = "https://www.apple.com/legal/privacy/en-ww/"

# Send a GET request to the website
response = requests.get(url)

# Create a BeautifulSoup object with the response content
soup = BeautifulSoup(response.content, "html.parser")

# Find the target parent element with CSS selector 'main.main[role="main"]'
parent_element = soup.select_one('main.main[role="main"]')

# Find all "h1", "p", and "li" elements inside the parent element
h1_elements = parent_element.select('h1')
p_elements = parent_element.select('p')
li_elements = parent_element.select('li')

# Extract the text content from each element
scraped_data = [element.get_text(strip=True) for element in h1_elements + p_elements + li_elements]

# Print the scraped data
for item in scraped_data:
    print(item)

# Save the scraped data into a CSV file
with open('apple_privacy_2023.csv', 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)

    # Write the data to the CSV file
    writer.writerow(['Scraped Data'])
    writer.writerows([[item] for item in scraped_data])
```
